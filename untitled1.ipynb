# Task
Analyze patient treatment and outcome data from "/content/outlook.csv", "/content/patients.csv", and "/content/treatment.csv". Merge the datasets, group by treatment type to find average recovery days and success rate, and pivot to show total cost per treatment type for each patient. Answer the following questions: Which treatment type has the highest average cost? Which patient had the longest recovery time? How would you analyze the success rate by gender?

## Load the datasets

### Subtask:
Load the three datasets into pandas DataFrames.


**Reasoning**:
Import pandas and load the three CSV files into dataframes as instructed.



```python
import pandas as pd

outlook_df = pd.read_csv('/content/outlook.csv')
patients_df = pd.read_csv('/content/patients.csv')
treatment_df = pd.read_csv('/content/treatment.csv')
```

## Merge the datasets

### Subtask:
Merge the three DataFrames based on common columns to create a unified dataset.


**Reasoning**:
Merge the three dataframes to create a unified dataset and display the head and columns of the merged dataframe to verify the merge.



```python
merged_df = pd.merge(outlook_df, treatment_df, on='treatment_id')
merged_df = pd.merge(merged_df, patients_df, on='treatment_id')
display(merged_df.head())
display(merged_df.columns)
```

## Analyze by treatment type

### Subtask:
Group the merged data by `treatment_type` and calculate the average recovery days and success rate for each type.


**Reasoning**:
Group the merged dataframe by `treatment_type` and calculate the mean of 'recovery_days_x' and the mean of 'success_x' after mapping 'Yes' to 1 and 'No' to 0.



```python
treatment_summary = merged_df.groupby('treatment_type').agg(
    avg_recovery_days=('recovery_days_x', 'mean'),
    success_rate=('success_x', lambda x: x.map({'Yes': 1, 'No': 0}).mean())
)
display(treatment_summary)
```

## Pivot for total cost

### Subtask:
Pivot the data to show each patient's total cost per treatment type.


**Reasoning**:
Create a pivot table to show the total cost per treatment type for each patient.



```python
patient_cost_pivot = merged_df.pivot_table(index='patient_id', columns='treatment_type', values='cost', aggfunc='sum')
display(patient_cost_pivot)
```
